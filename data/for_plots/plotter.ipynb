```python
import pandas as pd
import numpy as np
import os
import matplotlib
```

```python
dir = os.getcwd()
path = f'{dir}/Unified'
os.chdir(path)
```

```python
test_df = pd.read_csv('Stage_1.csv')
```

```python
df_1 = test_df[['i-uni-resnet-18-l1_stage_1 - val.overall_acc', 'i-uni-vgg16-l1_stage_1 - val.overall_acc', 'i-uni-convnext-l1_stage_1 - val.overall_acc']]
df_1 = df_1.rename(columns={'i-uni-resnet-18-l1_stage_1 - val.overall_acc': 'ResNet', 'i-uni-vgg16-l1_stage_1 - val.overall_acc': "VGG-16", 'i-uni-convnext-l1_stage_1 - val.overall_acc': "ConvNext"})
df_1 = df_1 * 100
```
